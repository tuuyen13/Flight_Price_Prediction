=>A chart clearly shows the variation in flight ticket prices depending on the number of days remaining until the departure date.
 
=> A common customer psychology when purchasing flight tickets is that the earlier they buy, the cheaper the ticket, or the closer to the departure date, the more expensive the ticket. To investigate this, the team has created a chart that illustrates the detailed flight ticket prices based on the number of days remaining until the departure date for each airline.

⇒ Within the range of 20-50 days before booking, prices increase slowly and remain stable.
As the remaining time before departure decreases to 20 days, prices start to rise sharply, reaching their peak at 2 days. To secure lower prices, it is advisable to book tickets at least 20 days before the travel date.

⇒ One day before the flight, ticket prices decrease threefold compared to the previous day. This can be explained by airlines flexibly adjusting ticket prices based on the actual situation of each specific flight to fill up empty seats.

# Build machine learning model

```python
# The dependent variable is the ticket price ('price'), while the independent variables are the remaining columns in the dataset.
x = df.drop(['price'], axis=1)
y = df['price']
```

## Train, test split

```python
# Split the dataset into Train and Test sets using the Train_test_split library. The ratio of Train set to Test set is 80:20.
# import libaries
from sklearn.model_selection import train_test_split
# train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Feature importance

## Choose models

The team executed three models in sequence:  DecisionTreeRegressor, RandomForestRegressor, KNeighborsRegressor

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
DRmodel = DecisionTreeRegressor()
RDmodel = RandomForestRegressor()
KNmodel = KNeighborsRegressor()
```

## Models evaluations

The experiments were conducted on three models, and the evaluation was performed using the following metrics: MAE (Mean Absolute Error), MSE (Mean Squared Error), RMSE (Root Mean Squared Error), R2 Score, RMSLE (Root Mean Squared Logarithmic Error), MAPE (Mean Absolute Percentage Error), and Adjusted R Square

```python
a={'Model Name':[], 'Mean_Absolute_Error_MAE':[] ,'Adj_R_Square':[] ,'Root_Mean_Squared_Error_RMSE':[] ,'Mean_Absolute_Percentage_Error_MAPE':[] ,'Mean_Squared_Error_MSE':[] ,'Root_Mean_Squared_Log_Error_RMSLE':[] ,'R2_score':[]}
Results=pd.DataFrame(a)
Results.head()
```

```python
MD = [DRmodel, RDmodel, KNmodel]

for models in MD:
    
    # Fit the model with train data
    
    models.fit(x_train, y_train)
    
    # Predict the model with test data

    y_pred = models.predict(x_test)
    
    # Print the model name
    
    print('Model Name: ', models)
    # Evaluation metrics for Regression analysis

    from sklearn import metrics

    print('Mean Absolute Error (MAE):', round(metrics.mean_absolute_error(y_test, y_pred),3))  
    print('Mean Squared Error (MSE):', round(metrics.mean_squared_error(y_test, y_pred),3))  
    print('Root Mean Squared Error (RMSE):', round(np.sqrt(metrics.mean_squared_error(y_test, y_pred)),3))
    print('R2_score:', round(metrics.r2_score(y_test, y_pred),6))
    print('Root Mean Squared Log Error (RMSLE):', round(np.log(np.sqrt(metrics.mean_squared_error(y_test, y_pred))),3))
    
    # Define the function to calculate the MAPE - Mean Absolute Percentage Error

    def MAPE (y_test, y_pred):
        y_test, y_pred = np.array(y_test), np.array(y_pred)
        return np.mean(np.abs((y_test - y_pred) / y_test)) * 100
    # Evaluation of MAPE 

    result = MAPE(y_test, y_pred)
    print('Mean Absolute Percentage Error (MAPE):', round(result, 2), '%')
    
    # Calculate Adjusted R squared values 

    r_squared = round(metrics.r2_score(y_test, y_pred),6)
    adjusted_r_squared = round(1 - (1-r_squared)*(len(y)-1)/(len(y)-x.shape[1]-1),6)
    print('Adj R Square: ', adjusted_r_squared)
    print('------------------------------------------------------------------------------------------------------------')
    #-------------------------------------------------------------------------------------------
    new_row = {'Model Name' : models,
               'Mean_Absolute_Error_MAE' : metrics.mean_absolute_error(y_test, y_pred),
               'Adj_R_Square' : adjusted_r_squared,
               'Root_Mean_Squared_Error_RMSE' : np.sqrt(metrics.mean_squared_error(y_test, y_pred)),
               'Mean_Absolute_Percentage_Error_MAPE' : result,
               'Mean_Squared_Error_MSE' : metrics.mean_squared_error(y_test, y_pred),
               'Root_Mean_Squared_Log_Error_RMSLE': np.log(np.sqrt(metrics.mean_squared_error(y_test, y_pred))),
               'R2_score' : metrics.r2_score(y_test, y_pred)}
    Results = Results.append(new_row, ignore_index=True)
    #------------------------------------------------------------
    
```

```python
results = Results.sort_values(by=['Adj_R_Square'], ascending=False)
results
```

Comparing the metrics, the team observed that Random Forest Regression yielded better results compared to the other models.
